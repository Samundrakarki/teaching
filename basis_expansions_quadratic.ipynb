```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt

mu = 0.0
sigma = 1.5
N= 31
k=3

X_train = np.linspace(-4,4,N)
X_train = X_train.reshape(-1,1)

y_train = np.zeros(N)
y_train = y_train.reshape(-1,1)

epsilon_train = np.random.normal(mu,sigma,N)
epsilon_train = epsilon_train.reshape(-1,1)

y_train = X_train*X_train + epsilon_train

poly = PolynomialFeatures(degree=2)
phi_X_train = poly.fit_transform(X_train)

neigh = LinearRegression()
neigh.fit(phi_X_train, y_train)

X_eval = np.linspace(-4,4,1000)
X_eval = X_eval.reshape(-1,1)
phi_X_eval = poly.fit_transform(X_eval)


plt.figure()
plt.plot(X_eval,neigh.predict(phi_X_eval), label="regression predictor")
plt.plot(X_eval,X_eval*X_eval, label="exact model")
plt.plot(X_train,y_train, 'rs', markersize=6, label="training set")
plt.legend()
plt.title("Regression by quadratic polynomial model")
plt.show()



```
