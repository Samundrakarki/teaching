```python
import numpy as np
import matplotlib.pyplot as plt

X = np.loadtxt("train.3",delimiter=",")

fig, axs = plt.subplots(1, 4,sharex=True, sharey=True)
for i in range(4):
    axs[i].imshow(-X[i,:].reshape((16,16)), cmap="gray")
```

```python
U,D,V_transp = np.linalg.svd(X,full_matrices=False)
V=V_transp.T

for d in [256,200,150,100,50]:
    print("d=%d"%(d))
    V_d = V[:,:d]
    x=[]
    x_tilde=[]
    fig, axs = plt.subplots(1, 4,sharex=True, sharey=True)
    for i in range(4):
        x_tmp=V_d @ V_d.T @ X[i,:]
        axs[i].imshow(-x_tmp.reshape((16,16)), cmap="gray")
        axs[i].set_title("d=%d"%(d))

```
